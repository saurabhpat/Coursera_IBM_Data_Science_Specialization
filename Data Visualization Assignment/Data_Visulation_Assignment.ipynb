```python
#Importing pandas and numpy libraries
import pandas as pd
import numpy as np
import matplotlib as mlt
import matplotlib.pyplot as plt
```

```python
df_ds = pd.read_csv('https://cocl.us/datascience_survey_data',index_col = 0)
print('Data Science Topic Survey Data Downloaded..!')
```

```python
df_ds
```

```python
df_ds.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)
df_ds_pct = ((df_ds / 2233) * 100).round(2)
```

```python
ax = df_ds_pct.plot(kind='bar',
                       figsize = (20, 8),
                       width = 0.8,
                       color = ['#5cb85c', '#5bc0de', '#d9534f'],
                       fontsize = 14)

plt.title('Percentage of Respondents Interests in Data Science Areas', fontsize=16)

for p in ax.patches:
    ax.annotate(str(p.get_height()) + '%', (p.get_x() * 1.005, p.get_height() * 1.03))
    
plt.show()
```
